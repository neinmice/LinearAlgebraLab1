```python
from typing import List

def is_palindrome(n):
    '''
    проверяем является ли число палиндромом
    важно!! делаем не через строки, а путём преобразований разрядов
    потому что скорость ресурсы и всё такое
    '''
    if n == 0:
        return 0
    buff = []
    length = 0
    while n >= 1:
        buff.append(n % 10)
        n //= 10
        length += 1
    fl = 1
    for i in range(length // 2):
        if buff[i] != buff[length - i - 1]:
            fl = 0
    if fl:
        return 1
    return 0
            
            

def make_palindromes(a0, a1, a2):
    '''
    генерировать палиндромы длины <= 5 с помощью
    всего трёх вводных ровно в сто раз
    эффективнее, чем перебирать все 100.000 значений
    '''
    if a0 == 0:
        if a1 == 0:
            pal1, pal2 = a2, a2 * 10 + a2
        else:
            pal1, pal2 = a1 * 100 + a2 * 10 + a1, a1 * 1000 + a2 * 100 + a2 * 10 + a1
    else:
        pal1, pal2 = a0 * 10000 + a1 * 1000 + a2 * 100 + a1 * 10 + a0, 0 # тут второй палиндром не генерируем, тк он будет > 100.000
    return pal1, pal2


def is_prime(n):
    '''
    Проверяет, является ли число простым.
    '''
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True


def palindromic_cubes_and_palindromic_primes() -> tuple[List[int], List[int]]:
    """
    Возвращает:
    tuple:
    - список всех палиндромов a < 100000, для которых a^3 — палиндром;
    - список всех простых p <= 10000, которые являются палиндромами.
    """


    valid1 = []
    valid2 = []
    
    for a0 in range(10):
        for a1 in range(10):
            for a2 in range(10):
                pal1, pal2 = make_palindromes(a0, a1, a2)
                sq1, sq2 = pal1 ** 3, pal2 ** 3
                if is_palindrome(sq1):
                    valid1.append(pal1)
                if is_palindrome(sq2):
                    valid1.append(pal2)
                if is_prime(pal1):
                    valid2.append(pal1)
                if is_prime(pal2):
                    valid2.append(pal2)

    return valid1, valid2
                    
                

palindromic_cubes_and_palindromic_primes()
                
    
```
