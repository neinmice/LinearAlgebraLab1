```python
from typing import List

def is_palindrome(n):
    '''
    проверяем является ли число палиндромом
    важно!! делаем не через строки, а путём преобразований разрядов
    потому что скорость ресурсы и всё такое
    '''
    if n == 0:
        return 0
    buff = []
    length = 0
    while n >= 1:
        buff.append(n % 10)
        n //= 10
        length += 1
    fl = 1
    for i in range(length // 2):
        if buff[i] != buff[length - i - 1]:
            fl = 0
    if fl:
        return 1
    return 0
            
            

def make_palindromes(a0, a1, a2):
    '''
    генерировать палиндромы длины <= 5 с помощью
    всего трёх вводных ровно в сто раз
    эффективнее, чем перебирать все 100.000 значений
    '''
    if a0 == 0:
        if a1 == 0:
            pal1, pal2 = a2, a2 * 10 + a2
        else:
            pal1, pal2 = a1 * 100 + a2 * 10 + a1, a1 * 1000 + a2 * 100 + a2 * 10 + a1
    else:
        pal1, pal2 = a0 * 10000 + a1 * 1000 + a2 * 100 + a1 * 10 + a0, 0 # тут второй палиндром не генерируем, тк он будет > 100.000
    return pal1, pal2

def erathosphene(n):
    '''
    классическая реализация решета Эратосфена
    '''
    if n < 2:
        return []

    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False 
    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]:
            for j in range(i * i, n + 1, i):
                is_prime[j] = False
    
    primes = [i for i in range(2, n + 1) if is_prime[i]]
    return primes

def is_prime(n):
    '''
    Проверяет, является ли число простым.
    '''
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def cyclic_permutations(n):
    '''
    циклические перестановки числа n
    '''
    if n < 10:
        return [n]
    
    s = str(n)
    length = len(s)
    permutations = []
    
    for i in range(length):
        # циклически сдвигаем строку
        perm = s[i:] + s[:i]
        permutations.append(int(perm))
    
    return permutations

def palindromic_squares_and_circular_primes() -> tuple[List[int], List[int]]:
    """
    Возвращает:
    tuple:
    - список всех палиндромов a < 100.000, для которых a^2 — палиндром;
    - список всех простых p < 1.000.000, все циклические перестановки цифр которых
    ,→ просты.
    """
    valid = []
    
    for a0 in range(10):
        for a1 in range(10):
            for a2 in range(10):
                pal1, pal2 = make_palindromes(a0, a1, a2)
                sq1, sq2 = pal1 ** 2, pal2 ** 2
                if is_palindrome(sq1):
                    valid.append(pal1)
                if is_palindrome(sq2):
                    valid.append(pal2)
    valid_primes = []
    # all_primes = erathosphene(1000000)
    # for prime in all_primes:
    #     perms = cyclic_permutations(prime)
    #     fl = 1
    #     for p in perms:
    #         if not is_prime(p):
    #             fl = 0
    #     if fl:
    #         valid_primes.extend(perms)
    # result = (valid, valid_primes)
    circular_primes = []
    for p in range(2, 1000000):
        if is_prime(p):
            permutations = cyclic_permutations(p)
            all_prime = True
            
            for perm in permutations:
                if not is_prime(perm):
                    all_prime = False
                    break
            
            if all_prime:
                circular_primes.append(p)
    

    return valid, circular_primes
                    
                

palindromic_squares_and_circular_primes()
                
    
```
